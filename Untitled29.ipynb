```python
import pandas as pd
import ta  # Technical analysis library
from ta.volatility import BollingerBands
from ta.trend import MACD, SMAIndicator, EMAIndicator
from ta.momentum import RSIIndicator

# üîπ Load your data (already uploaded)
bitcoin = pd.read_csv("Bitcoin_data.csv")
tesla = pd.read_csv("Tesla_stock.csv")

# üîπ Convert to datetime
bitcoin['Date'] = pd.to_datetime(bitcoin['Date'], dayfirst=True)
tesla['Date'] = pd.to_datetime(tesla['Date'], dayfirst=True)

# üîπ Set index
bitcoin.set_index('Date', inplace=True)
tesla.set_index('Date', inplace=True)

def generate_features(df):
    df = df.copy()
    df['log_close'] = np.log(df['Close'])
    df['returns'] = df['Close'].pct_change()

    # ‚ûï SMA
    for win in [12, 20, 26, 48, 220]:
        df[f'SMA{win}'] = SMAIndicator(df['Close'], window=win).sma_indicator()

    # ‚ûï EMA
    for win in [12, 26, 48, 220]:
        df[f'EMA{win}'] = EMAIndicator(df['Close'], window=win).ema_indicator()

    # ‚ûï Bollinger Bands
    bb = BollingerBands(close=df['Close'], window=20, window_dev=2)
    df['bb_upper'] = bb.bollinger_hband()
    df['bb_lower'] = bb.bollinger_lband()

    # ‚ûï RSI
    df['RSI'] = RSIIndicator(close=df['Close'], window=14).rsi()

    # ‚ûï MACD
    macd = MACD(close=df['Close'])
    df['MACD'] = macd.macd()

    df.dropna(inplace=True)
    return df

# Generate for both assets
bitcoin_feat = generate_features(bitcoin)
tesla_feat = generate_features(tesla)

# üíæ Optional: Save to CSV
bitcoin_feat.to_csv("bitcoin_features.csv")
tesla_feat.to_csv("tesla_features.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Function to prepare LSTM sequences
def create_lstm_data(series, window=30):
    X, y = [], []
    for i in range(window, len(series)):
        X.append(series[i-window:i])
        y.append(series[i])
    return np.array(X), np.array(y)

# Period definitions
periods = {
    "Period 1": ('2016-01-01', '2019-08-31'),
    "Period 2": ('2019-09-01', '2023-03-31'),
    "Period 3": ('2016-01-01', '2023-03-31'),
}

# Asset dictionary
assets = {
    "Bitcoin": bitcoin_feat,  # Assumes 'log_close' already exists
    "Tesla": tesla_feat
}

# Dictionary to store LSTM-ready data
lstm_data = {}

# Loop through assets and periods
for asset_name, df in assets.items():
    lstm_data[asset_name] = {}

    for label, (start, end) in periods.items():
        subset = df.loc[start:end].copy().dropna()

        # Normalize the log_close for LSTM
        scaler = MinMaxScaler()
        scaled_close = scaler.fit_transform(subset[['log_close']])

        # Create LSTM sequences
        X, y = create_lstm_data(scaled_close, window=30)

        lstm_data[asset_name][label] = {
            'X': X,
            'y': y,
            'scaler': scaler,
            'dates': subset.index[30:]
        }

print("‚úÖ LSTM data prepared for Bitcoin and Tesla.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import numpy as np

# üîπ Get Bitcoin Period 1 data
X = lstm_data['Bitcoin']['Period 1']['X']
y = lstm_data['Bitcoin']['Period 1']['y']
scaler = lstm_data['Bitcoin']['Period 1']['scaler']
dates = lstm_data['Bitcoin']['Period 1']['dates']

# üîπ Train/Test Split (80/20)
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]
test_dates = dates[split_idx:]

# üîπ Build LSTM model
model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(X.shape[1], 1)),
    Dropout(0.2),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse')

# üîπ Train
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=0)

# üîπ Predict
y_pred = model.predict(X_test)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
y_pred_inv = scaler.inverse_transform(y_pred)

# üîπ Evaluation
rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
mae = mean_absolute_error(y_test_inv, y_pred_inv)
print(f"üìä Bitcoin Period 1 LSTM: RMSE = {rmse:.2f}, MAE = {mae:.2f}")

# üîπ Plot
plt.figure(figsize=(14,5))
plt.plot(test_dates, y_test_inv, label='Actual', color='blue')
plt.plot(test_dates, y_pred_inv, label='Predicted', color='orange', linestyle='--')
plt.title("Bitcoin Period 1 ‚Äì LSTM Forecast (Close Price)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# üìö Required Libraries
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from google.colab import files

# üõ†Ô∏è Helper Functions
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length, :-1])
        y.append(data[i+seq_length, -1])
    return np.array(X), np.array(y)

def evaluate_model(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = 1 - (np.sum((y_true - y_pred)**2) / np.sum((y_true - np.mean(y_true))**2))
    return rmse, mse, mae, mape, r2

def build_model(model_type, input_shape):
    model = Sequential()
    if model_type == "Vanilla":
        model.add(LSTM(50, activation='relu', input_shape=input_shape))
    elif model_type == "Stacked 1":
        model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=input_shape))
        model.add(LSTM(50, activation='relu'))
    elif model_type == "Stacked 2":
        model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=input_shape))
        model.add(LSTM(50, activation='relu', return_sequences=True))
        model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# ‚úÖ Feature Sets with Precise Names
feature_sets = {
    "Close": ["Close"],
    "Close, Returns": ["Close", "returns"],
    "Close, EMA12, EMA26, EMA48, EMA220": ["Close", "EMA12", "EMA26", "EMA48", "EMA220"],
    "Close, SMA12, SMA26, SMA20, SMA48, SMA220": ["Close", "SMA12", "SMA26", "SMA20", "SMA48", "SMA220"],
    "Close, RSI, MACD": ["Close", "RSI", "MACD"],
    "Close, Upper Bollinger Band, Lower Bollinger Band": ["Close", "Upper_BB", "Lower_BB"],
    "Close, EMA + SMA + Bollinger": [
        "Close", "EMA12", "EMA26", "EMA48", "EMA220",
        "SMA12", "SMA26", "SMA20", "SMA48", "SMA220",
        "Upper_BB", "Lower_BB"
    ],
    "Close, EMA + SMA + Bollinger + RSI + MACD": [
        "Close", "EMA12", "EMA26", "EMA48", "EMA220",
        "SMA12", "SMA26", "SMA20", "SMA48", "SMA220",
        "Upper_BB", "Lower_BB", "RSI", "MACD"
    ]
}

# ‚úÖ Periods
periods = {
    "Period 1": ('2016-01-01', '2019-08-31'),
    "Period 2": ('2019-09-01', '2023-03-31'),
    "Period 3": ('2016-01-01', '2023-03-31')
}

# üì• Load data
btc = pd.read_csv("bitcoin_features.csv", parse_dates=['Date'])
btc.set_index("Date", inplace=True)
btc["returns"] = btc["Close"].pct_change().fillna(0)

# üß™ Train + Evaluate
seq_len = 10
model_types = ["Vanilla", "Stacked 1", "Stacked 2"]
all_results = []

for feat_name, features in feature_sets.items():
    for period_name, (start, end) in periods.items():
        best_result = None
        best_rmse = float("inf")
        for model_type in model_types:
            print(f"üîÅ {feat_name} | {period_name} | {model_type}")
            df = btc.loc[start:end, features].dropna().copy()
            df["Target"] = df["Close"]

            scaler = MinMaxScaler()
            scaled = scaler.fit_transform(df)
            scaled_df = pd.DataFrame(scaled, columns=features + ["Target"], index=df.index)

            n = len(scaled_df)
            train_size = int(n * 0.8)
            train = scaled_df.iloc[:train_size].values
            test = scaled_df.iloc[train_size:].values

            X_train, y_train = create_sequences(train, seq_len)
            X_test, y_test = create_sequences(test, seq_len)

            model = build_model(model_type, (X_train.shape[1], X_train.shape[2]))
            model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

            y_pred = model.predict(X_test, verbose=0)
            y_true = y_test

            # Inverse scale
            y_pred_rescaled = scaler.inverse_transform(
                np.concatenate((np.zeros((len(y_pred), len(features))), y_pred), axis=1)
            )[:, -1]
            y_true_rescaled = scaler.inverse_transform(
                np.concatenate((np.zeros((len(y_true), len(features))), y_true.reshape(-1, 1)), axis=1)
            )[:, -1]

            rmse, mse, mae, mape, r2 = evaluate_model(y_true_rescaled, y_pred_rescaled)

            if rmse < best_rmse:
                best_rmse = rmse
                best_result = [feat_name, period_name, model_type, rmse, mse, mae, f"{mape:.2f}%", round(r2, 4)]

        all_results.append(best_result)

# üßæ Save Final Best Results Table
summary_df = pd.DataFrame(all_results, columns=["Features", "Period", "Best Model", "RMSE", "MSE", "MAE", "MAPE", "R2"])
summary_df.to_csv("bitcoin_lstm_multivariate_summary_best.csv", index=False)
files.download("bitcoin_lstm_multivariate_summary_best.csv")
print("‚úÖ Final table generated and downloaded.")
```

```python
# üìö Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# üóÇÔ∏è Load and preprocess Bitcoin data
btc = pd.read_csv("Bitcoin_data.csv", parse_dates=["Date"])
btc.set_index("Date", inplace=True)

# üìÖ Periods
periods = {
    "Period 1": ("2016-01-01", "2019-08-31"),
    "Period 2": ("2019-09-01", "2023-03-31"),
    "Period 3": ("2016-01-01", "2023-03-31")
}

# üì¶ Loop over each period
for label, (start, end) in periods.items():
    print(f"üöÄ Running LSTM on Differenced Data for {label}")

    # Slice and compute differenced Close
    df = btc.loc[start:end].copy()
    df["diff_close"] = df["Close"].diff()
    df.dropna(inplace=True)

    # Normalize differenced values
    scaler = MinMaxScaler()
    df["scaled_diff"] = scaler.fit_transform(df[["diff_close"]])

    # Sequence builder
    def create_sequences(series, seq_len=10):
        X, y = [], []
        for i in range(len(series) - seq_len):
            X.append(series[i:i+seq_len])
            y.append(series[i+seq_len])
        return np.array(X), np.array(y)

    seq_len = 10
    X, y = create_sequences(df["scaled_diff"].values.reshape(-1, 1), seq_len)
    dates = df.index[seq_len:]

    # Split into training and test
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]
    test_dates = dates[split:]

    # Build LSTM model
    model = Sequential()
    model.add(LSTM(64, input_shape=(seq_len, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=0)

    # Predict
    y_pred = model.predict(X_test)
    y_pred_inv = scaler.inverse_transform(y_pred)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

    # üîπ PLOT 1: Differenced Values
    plt.figure(figsize=(12,5))
    plt.plot(test_dates, y_test_inv, label="Actual Differenced Values", color='blue')
    plt.plot(test_dates, y_pred_inv, label="Predicted Differenced Values", color='magenta')
    plt.title(f"BTC Predicted 1st Derivative Values ‚Äì {label}")
    plt.xlabel("Time")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"btc_diff_{label.replace(' ', '_')}.png", dpi=300)
    plt.show()

    # üîÅ Reconstruct Close Prices from differences
    last_train_close = df["Close"].iloc[split + seq_len - 1]
    pred_close = np.r_[last_train_close, y_pred_inv.flatten()].cumsum()[1:]
    actual_close = np.r_[last_train_close, y_test_inv.flatten()].cumsum()[1:]

    # üîπ PLOT 2: Reconstructed Close Prices
    plt.figure(figsize=(12,5))
    plt.plot(test_dates, actual_close, label="Actual Close Price", color='blue')
    plt.plot(test_dates, pred_close, label="Predicted Close Price", color='magenta')
    plt.title(f"Actual vs. Predicted Close Prices ‚Äì {label}")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"btc_diff_reconstructed_{label.replace(' ', '_')}.png", dpi=300)
    plt.show()
```

```python
# üìö Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# üì• Load Bitcoin dataset
btc = pd.read_csv("Bitcoin_data.csv", parse_dates=["Date"])
btc.set_index("Date", inplace=True)

# üìÖ Periods
periods = {
    "Period 2": ("2019-09-01", "2023-03-31"),
    "Period 3": ("2016-01-01", "2023-03-31")
}

# ‚úÖ Sequence builder
def create_sequences(series, seq_len=10):
    X, y = [], []
    for i in range(len(series) - seq_len):
        X.append(series[i:i+seq_len])
        y.append(series[i+seq_len])
    return np.array(X), np.array(y)

# üîÅ Loop over both periods
for label, (start, end) in periods.items():
    print(f"\nüöÄ Processing {label} ‚Äî Direct vs Differenced LSTM")

    df = btc[start:end].copy()

    # ---------------------
    # üß† 1. Direct LSTM
    # ---------------------
    scaler_direct = MinMaxScaler()
    df["scaled_close"] = scaler_direct.fit_transform(df[["Close"]])

    X_d, y_d = create_sequences(df["scaled_close"].values.reshape(-1, 1))
    dates_d = df.index[10:]

    split = int(len(X_d) * 0.8)
    X_train_d, X_test_d = X_d[:split], X_d[split:]
    y_train_d, y_test_d = y_d[:split], y_d[split:]
    test_dates_d = dates_d[split:]

    model_d = Sequential([
        LSTM(64, input_shape=(X_d.shape[1], 1)),
        Dense(1)
    ])
    model_d.compile(optimizer='adam', loss='mse')
    model_d.fit(X_train_d, y_train_d, epochs=30, batch_size=32, validation_split=0.1, verbose=0)

    y_pred_d = model_d.predict(X_test_d)
    y_test_inv_direct = scaler_direct.inverse_transform(y_test_d.reshape(-1, 1))
    y_pred_inv_direct = scaler_direct.inverse_transform(y_pred_d)

    # ---------------------
    # üß† 2. Differenced LSTM
    # ---------------------
    df["diff_close"] = df["Close"].diff()
    df.dropna(inplace=True)

    scaler_diff = MinMaxScaler()
    df["scaled_diff"] = scaler_diff.fit_transform(df[["diff_close"]])

    X_diff, y_diff = create_sequences(df["scaled_diff"].values.reshape(-1, 1))
    dates_diff = df.index[10:]

    split_diff = int(len(X_diff) * 0.8)
    X_train_diff, X_test_diff = X_diff[:split_diff], X_diff[split_diff:]
    y_train_diff, y_test_diff = y_diff[:split_diff], y_diff[split_diff:]
    test_dates_diff = dates_diff[split_diff:]

    model_diff = Sequential([
        LSTM(64, input_shape=(X_diff.shape[1], 1)),
        Dense(1)
    ])
    model_diff.compile(optimizer='adam', loss='mse')
    model_diff.fit(X_train_diff, y_train_diff, epochs=30, batch_size=32, validation_split=0.1, verbose=0)

    y_pred_diff = model_diff.predict(X_test_diff)
    y_pred_inv_diff = scaler_diff.inverse_transform(y_pred_diff)
    y_test_inv_diff = scaler_diff.inverse_transform(y_test_diff.reshape(-1, 1))

    # üîÅ Reconstruct Close from differenced predictions
    last_real_close = df["Close"].iloc[split_diff + 10 - 1]
    reconstructed_pred = np.r_[last_real_close, y_pred_inv_diff.flatten()].cumsum()[1:]
    reconstructed_actual = np.r_[last_real_close, y_test_inv_diff.flatten()].cumsum()[1:]

    # ---------------------
    # üìà Plot Overlay
    # ---------------------
    plt.figure(figsize=(14, 6))
    plt.plot(test_dates_d, y_test_inv_direct, label="Actual Close (Direct)", color='black')
    plt.plot(test_dates_d, y_pred_inv_direct, label="Predicted Close (Direct)", color='blue', linestyle='--')
    plt.plot(test_dates_diff, reconstructed_actual, label="Actual Close (Diff)", color='green')
    plt.plot(test_dates_diff, reconstructed_pred, label="Predicted Close (Diff)", color='magenta', linestyle=':')

    plt.title(f"Overlay: Direct vs Differenced LSTM Close Price Prediction ‚Äì {label}")
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"overlay_direct_vs_diff_{label.replace(' ', '_')}.png", dpi=300)
    plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import pandas as pd
import numpy as np

btc = pd.read_csv("Bitcoin_data.csv", parse_dates=["Date"])
btc.set_index("Date", inplace=True)

# Define periods
periods = {
    "Period 1": ("2016-01-01", "2019-08-31"),
    "Period 2": ("2019-09-01", "2023-03-31"),
    "Period 3": ("2016-01-01", "2023-03-31")
}

# Container for storing results
y_test_inv_diff_p1 = y_pred_inv_diff_p1 = None
y_test_inv_diff_p2 = y_pred_inv_diff_p2 = None
y_test_inv_diff_p3 = y_pred_inv_diff_p3 = None

def create_sequences(data, seq_len=10):
    X, y = [], []
    for i in range(len(data) - seq_len):
        X.append(data[i:i+seq_len])
        y.append(data[i+seq_len])
    return np.array(X), np.array(y)

# Train LSTM on differenced data for all periods
for period_name, (start, end) in periods.items():
    df = btc[start:end].copy()
    df["diff_close"] = df["Close"].diff()
    df.dropna(inplace=True)

    scaler = MinMaxScaler()
    df["scaled_diff"] = scaler.fit_transform(df[["diff_close"]])

    X, y = create_sequences(df["scaled_diff"].values.reshape(-1, 1))
    split = int(len(X) * 0.8)
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    model = Sequential([
        LSTM(64, input_shape=(X.shape[1], 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.1, verbose=0)

    y_pred = model.predict(X_test)
    y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))
    y_pred_inv = scaler.inverse_transform(y_pred)

    if period_name == "Period 1":
        y_test_inv_diff_p1 = y_test_inv
        y_pred_inv_diff_p1 = y_pred_inv
    elif period_name == "Period 2":
        y_test_inv_diff_p2 = y_test_inv
        y_pred_inv_diff_p2 = y_pred_inv
    elif period_name == "Period 3":
        y_test_inv_diff_p3 = y_test_inv
        y_pred_inv_diff_p3 = y_pred_inv
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd
import numpy as np

def evaluate_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / np.maximum(np.abs(y_true), 1e-10))) * 100
    r2 = r2_score(y_true, y_pred)
    return round(rmse, 2), round(mse, 2), round(mae, 2), f"{mape:.2f}%", round(r2, 2)

results = []

for period in ["Period 1", "Period 2", "Period 3"]:
    if period == "Period 1":
        y_true = y_test_inv_diff_p1
        y_pred = y_pred_inv_diff_p1
    elif period == "Period 2":
        y_true = y_test_inv_diff_p2
        y_pred = y_pred_inv_diff_p2
    elif period == "Period 3":
        y_true = y_test_inv_diff_p3
        y_pred = y_pred_inv_diff_p3

    rmse, mse, mae, mape, r2 = evaluate_metrics(y_true.flatten(), y_pred.flatten())
    results.append([period, rmse, mse, mae, mape, r2])

df_results = pd.DataFrame(results, columns=["Period", "RMSE", "MSE", "MAE", "MAPE", "R2"])
print(df_results.to_string(index=False))
df_results.to_csv("differenced_lstm_evaluation_summary.csv", index=False)
```
